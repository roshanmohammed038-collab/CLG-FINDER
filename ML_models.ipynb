```python
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import warnings
import pandas as pd
import seaborn as sns
```

```python
df1=sns.load_dataset('iris') #checking grid cv using knn
```

```python
df1.head()
```

```python
df1['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df1.drop('species',axis=1)
```

```python
y=df1['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model_knn=KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(X_train,y_train)
```

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(gamma='auto')
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test) # same as KNN
```

```python
model_knn.score(X_test,y_test) #model overfit coz 100% accuracy for n=5 bit n=11 its 98
```

```python
#similarlly  for SVC its the same accuracy its different for n diff parameters to overcome we use grid search hyper parameters!
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier=GridSearchCV((model_svm),{
    'C':[1,10,20,30], #penalty
    'kernel':['rbf','linear'], #2 std kernels

},cv=5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
res=pd.DataFrame(classifier.cv_results_)
```

```python
res
```

```python
res[['param_C','param_kernel','mean_test_score']] #  checking the values for  2 parametrs with higheest accuracy
```

```python
#if we keep the val of parameter_c :1 and kernel:rbf  and gamma :auto we get 98% accuracy
```
